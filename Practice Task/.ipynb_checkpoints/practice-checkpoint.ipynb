```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('data/Advertising.csv')
df.head()
```

```python
df = df.drop(columns =['Unnamed: 0'])
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
x = df.drop(columns=['Sales'])
x.head()
```

```python
y = df['Sales']
y.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
print(model.predict([[25, 11, 29.7]]))
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
pd.DataFrame({'y_test':y_test, 'y_predict':y_pred})
```

```python
print('MSE: ', mean_squared_error(y_test, y_pred))
print('RMSE: ', np.sqrt(mean_squared_error(y_test, y_pred)))      
print('R2 Score: ', r2_score(y_test, y_pred))      
```

```python
coefficeints = model.coef_
print(coefficeints)

intercept = model.intercept_
print(intercept)
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.scatter(x_train['TV'], y_train, alpha=0.7, edgecolors='b')
plt.title('TV vs Sales')
plt.xlabel('TV')
plt.ylabel('Sales')

plt.subplot(1, 3, 2)
plt.scatter(x_train['Radio'], y_train, alpha=0.7, edgecolors='b')
plt.title('Radio vs Sales')
plt.xlabel('Radio')
plt.ylabel('Sales')

plt.subplot(1, 3, 3)
plt.scatter(x_train['Newspaper'], y_train, alpha=0.7, edgecolors='b')
plt.title('Newspaper vs Sales')
plt.xlabel('Newspaper')
plt.ylabel('Sales')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 5))

# Coefficients and intercept from your trained model
coefficients = model.coef_  # Model's coefficients for each feature
intercept = model.intercept_  # Model's intercept

# Plot 1: TV vs Sales
plt.subplot(1, 3, 1)
plt.scatter(x_test['TV'], y_test, alpha=0.7, edgecolors='b')

# Plot the straight line y = mx + c
y_line_tv = coefficients[0] * x_test['TV'] + intercept
plt.plot(x_test['TV'], y_line_tv, color='r')

plt.title('TV vs Sales')
plt.xlabel('TV')
plt.ylabel('Sales')
plt.legend()

# Plot 2: Radio vs Sales
plt.subplot(1, 3, 2)
plt.scatter(x_test['Radio'], y_test, alpha=0.7, edgecolors='b')

# Plot the straight line y = mx + c
y_line_radio = coefficients[1] * x_test['Radio'] + intercept
plt.plot(x_test['Radio'], y_line_radio, color='r')

plt.title('Radio vs Sales')
plt.xlabel('Radio')
plt.ylabel('Sales')
plt.legend()

# Plot 3: Newspaper vs Sales
plt.subplot(1,3,3)
plt.scatter(x_test['Newspaper'], y_test, alpha=0.7, edgecolors='b')

# Plot the straight line y = mx + c
y_line_newspaper = coefficients[2] * x_test['Newspaper'] + intercept
plt.plot(x_test['Newspaper'], y_line_newspaper, color='r')

plt.title('Newspaper vs Sales')
plt.xlabel('Newspaper')
plt.ylabel('Sales')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Coefficients and intercept from your trained model
coefficients = model.coef_  # Model's coefficients for each feature
intercept = model.intercept_  # Model's intercept

# Scatter plot for all three features
plt.scatter(x_test['TV'], y_test, alpha=0.7, edgecolors='b', label='TV vs Sales', color='black')
plt.scatter(x_test['Radio'], y_test, alpha=0.7, edgecolors='b', label='Radio vs Sales', color='green')
plt.scatter(x_test['Newspaper'], y_test, alpha=0.7, edgecolors='b', label='Newspaper vs Sales', color='pink')

# Plot the straight lines y = mx + c for each feature
y_line_tv = coefficients[0] * x_test['TV'] + intercept
plt.plot(x_test['TV'], y_line_tv, color='red', label='TV Line')

y_line_radio = coefficients[1] * x_test['Radio'] + intercept
plt.plot(x_test['Radio'], y_line_radio, color='orange', label='Radio Line')

y_line_newspaper = coefficients[2] * x_test['Newspaper'] + intercept
plt.plot(x_test['Newspaper'], y_line_newspaper, color='yellow', label='Newspaper Line')

# Titles and labels
plt.title('TV, Radio, and Newspaper vs Sales')
plt.xlabel('Advertising Expenditure')
plt.ylabel('Sales')

# Show legend
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```
