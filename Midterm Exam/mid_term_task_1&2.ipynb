```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/Student Depression Dataset.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
most_frequent_city = df['City'].mode()[0]
df['City']= df['City'].fillna(most_frequent_city)
```

```python
df['Degree'] = df['Degree'].fillna('Others')
```

```python
df['Work/Study Hours'] = df['Work/Study Hours'].fillna(df['Work/Study Hours'].mean())
```

```python
df['Financial Stress'] = df['Financial Stress'].fillna(df['Financial Stress'].median())
```

```python
df['Family History of Mental Illness'] = df['Family History of Mental Illness'].fillna(df['Family History of Mental Illness'].mode()[0])
```

```python
df['Sleep Duration'] = df['Sleep Duration'].fillna('Others')
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
# pd.get_dummies(df, columns= ['Gender',, dtype=int)
```

```python
df = df.drop(['id','City', 'Profession', 'Sleep Duration', 'Degree'], axis=1)
```

```python
df.head()
```

```python
df.info()
```

```python
df = pd.get_dummies(df, columns=['Gender', 'Dietary Habits', 'Have you ever had suicidal thoughts ?', 'Family History of Mental Illness'], drop_first=True, dtype=int)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Age'] = (df['Age'] - df['Age'].min()) / (df['Age'].max() - df['Age'].min())
```

```python
min_cgpa = df['CGPA'].min()
max_cgpa = df['CGPA'].max()

df['CGPA'] = (df['CGPA'] - min_cgpa) / (max_cgpa - min_cgpa)
```

```python
min_satisfaction = df['Study Satisfaction'].min()
max_satisfaction = df['Study Satisfaction'].max()
df['Study Satisfaction'] = (df['Study Satisfaction'] - min_satisfaction) / (max_satisfaction - min_satisfaction)
```

```python
min_pressure = df['Work Pressure'].min()
max_pressure = df['Work Pressure'].max()
df['Work Pressure'] = (df['Work Pressure'] - min_pressure) / (max_pressure - min_pressure)
```

```python
min_job_satisfaction = df['Job Satisfaction'].min()
max_job_satisfaction = df['Job Satisfaction'].max()
df['Job Satisfaction'] = (df['Job Satisfaction'] - min_job_satisfaction) / (max_job_satisfaction - min_job_satisfaction)
```

```python
min_hours = df['Work/Study Hours'].min()
max_hours = df['Work/Study Hours'].max()
df['Work/Study Hours_normalized_minmax'] = (df['Work/Study Hours'] - min_hours) / (max_hours - min_hours)
```

```python
min_stress = df['Financial Stress'].min()
max_stress = df['Financial Stress'].max()
df['Financial Stress'] = (df['Financial Stress'] - min_stress) / (max_stress - min_stress)
```

```python
min_ap = df['Academic Pressure'].min()
max_ap = df['Academic Pressure'].max()
df['Academic Pressure'] = (df['Academic Pressure'] - min_ap) / (max_ap - min_ap)
```

```python
x = df.drop(columns=['Depression'])
x.head()
```

```python
y = df['Depression']
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(x_train, y_train)
```

```python
y_pred = knn.predict(x_test)
```

```python
pd.DataFrame({'y_test':y_test, 'y_pred':y_pred})
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Predicted 0', 'Predicted 1', 'Predicted 2'],
            yticklabels=['Actual 0', 'Actual 1', 'Actual 2'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

precision = precision_score(y_test, y_pred, average='weighted') # Use 'weighted' for multi-class
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
```

### Underfitting

A statistical model or a machine learning algorithm is said to have underfitting when a model is too simple to capture data complexities. It represents the inability of the model to learn the training data effectively result in poor performance both on the training and testing data. In simple terms, an underfit models are inaccurate, especially when applied to new, unseen examples. It mainly happens when we uses very simple model with overly simplified assumptions. To address underfitting problem of the model, we need to use more complex models, with enhanced feature representation.

### Overfitting

A statistical model is said to be overfitted when the model does not make accurate predictions on testing data.  Overfitting is a problem where the evaluation of machine learning algorithms on training data is different from unseen data.

### Bias

Bias refers to the error due to overly simplistic assumptions in the learning algorithm. These assumptions make the model easier to comprehend and learn but might not capture the underlying complexities of the data. It is the error due to the models inability to represent the true relationship between input and output accurately. When a model has poor performance both on the training and testing data means high bias because of the simple model, indicating underfitting.
