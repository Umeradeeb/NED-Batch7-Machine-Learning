```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/Placement_Data_Full_Class.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['ssc_p'] = df['ssc_p'].fillna(df['ssc_p'].mean())
```

```python
df['ssc_b'] = df['ssc_b'].fillna(df['ssc_b'].mode()[0])
```

```python
df['hsc_s'] = df['hsc_s'].fillna(df['hsc_s'].mode()[0])
```

```python
df['degree_p'] = df['degree_p'].fillna(df['degree_p'].mean())
```

```python
df['degree_t'] = df['degree_t'].fillna(df['degree_t'].mode()[0])
```

```python
df['workex'] = df['workex'].fillna(df['workex'].mode()[0])
```

```python
df['etest_p'] = df['etest_p'].fillna(df['etest_p'].mean())
```

```python
df['specialisation'] = df['specialisation'].fillna(df['specialisation'].mode()[0])
```

```python
df['salary'] = df['salary'].fillna(df['salary'].mean())
```

```python
df.isnull().sum()
```

for null values filling, I have used mode for object data type. For numerical columns, I used median and mean accordingly for skewed data with outliers and for balanced, symmetric data respectively. 

```python
df.info()
```

```python
df.head()
```

```python
df = df.drop(['sl_no','ssc_b','hsc_b','hsc_s'], axis=1)
```

```python
df.info()
```

```python
df = pd.get_dummies(df, columns=['gender','degree_t','workex','specialisation','status'], drop_first=True, dtype=int)
df.head()
```

```python
df.info()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['salary'], kde=True)
plt.title('Distribution of Salary')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
Q1 = df['salary'].quantile(0.25)
Q3 = df['salary'].quantile(0.75)
IQR = Q3 - Q1

# Define the upper and lower bounds for outliers
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

# Cap the outliers using the upper and lower bounds
df['salary'] = np.where(df['salary'] > upper_bound, upper_bound,
                       np.where(df['salary'] < lower_bound, lower_bound, df['salary']))

# Visualize the distribution of the salary column after capping outliers
plt.figure(figsize=(8, 6))
sns.histplot(df['salary'], kde=True)
plt.title('Distribution of Salary after Capping Outliers')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python

# Scatter plots
plt.figure(figsize=(10, 6))
sns.scatterplot(x='ssc_p', y='salary', data=df)
plt.title('Scatter Plot: ssc_p vs. Salary')
plt.xlabel('Secondary Education Percentage')
plt.ylabel('Salary')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='hsc_p', y='salary', data=df)
plt.title('Scatter Plot: hsc_p vs. Salary')
plt.xlabel('Higher Secondary Education Percentage')
plt.ylabel('Salary')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='degree_p', y='salary', data=df)
plt.title('Scatter Plot: degree_p vs. Salary')
plt.xlabel('Degree Percentage')
plt.ylabel('Salary')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='etest_p', y='salary', data=df)
plt.title('Scatter Plot: etest_p vs. Salary')
plt.xlabel('Employability Test Percentage')
plt.ylabel('Salary')
plt.show()


# Boxplots
plt.figure(figsize=(10, 6))
sns.boxplot(x='gender_M', y='salary', data=df)
plt.title('Box Plot: Gender vs. Salary')
plt.xlabel('Gender')
plt.ylabel('Salary')
plt.show()


plt.figure(figsize=(10, 6))
sns.boxplot(x='specialisation_Mkt&HR', y='salary', data=df)
plt.title('Box Plot: Specialization vs. Salary')
plt.xlabel('Specialization')
plt.ylabel('Salary')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='workex_Yes', y='salary', data=df)
plt.title('Box Plot: Work Experience vs. Salary')
plt.xlabel('Work Experience')
plt.ylabel('Salary')
plt.show()


# Correlation heatmap
plt.figure(figsize=(12, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

### Key Insights from the Dataset

#### 1. Salary Distribution:
The initial salary distribution is right-skewed, with a few high salary outliers. After capping these outliers using the IQR method, the distribution appears more normalized.

#### 2. Academic Performance and Salary:
Scatter plots reveal a positive correlation between academic performance (`ssc_p`, `hsc_p`, `degree_p`) and salary. Students with higher percentages in secondary, higher secondary, and degree education tend to have higher salaries. There's also a moderate correlation between the employability test percentage (`etest_p`) and salary.

#### 3. Gender and Salary:
The boxplot comparing gender and salary suggests a possible difference in average salaries between genders, with males tending to earn slightly higher on average. However, the difference is not substantial and could be influenced by other factors.

#### 4. Specialization and Salary:
The boxplot for specialization shows that Marketing and HR specialization has a slightly lower median salary compared to the other specializations. However, there is considerable overlap in salary ranges.

#### 5. Work Experience and Salary:
The boxplot indicates that candidates with prior work experience (`workex_Yes`) generally earn higher salaries compared to those without experience.

#### 6. Correlation Heatmap:

```python
x = df.drop(columns=['salary'])
x.head()
```

```python
y = df['salary']
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
pd.DataFrame({'y_test':y_test, 'y_predict':y_pred})
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error, r2_score
print('MAE:', mean_absolute_error(y_test, y_pred))
print('MSE:', mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(mean_squared_error(y_test, y_pred)))
print('R2_Score:', r2_score(y_test, y_pred))
```
